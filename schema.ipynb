```python
# create schema links using the format 
temp_train = train.head(100)
# print(tabulate(temp_train, headers='keys', tablefmt='pretty'))

ops = {0:'=', 1:'>', 2:'<', 3:'>=', 4:'<='}
for index, row in temp_train.iterrows():
    cur = row['format']
    
    # foreign keys
    fk = []
    if len(cur['table']) == 1:
        fk = None
    else:
        for i in range(0, len(cur['table'])):
            for j in range(i+1, len(cur['table'])):
                fk.append((tables["table_names"][0][i]+".HADM_ID", tables["table_names"][0][j]+".HADM_ID"))
    print("Foreign Keys: " + str(fk))
    
    # conditions
    conditions = []
    if len(cur['cond']) == 0:
        conditions = None
    else:
        for i in range(0, len(cur['cond'])):
            table = tables["table_names"][0][cur['cond'][i][0]]
            column_name = tables["column_names"][0][cur['cond'][i][1]][1]
            cond = ops[cur['cond'][i][2]]
            value = cur['cond'][i][3]
            conditions.append(table + "." + column_name + " " + cond + " " + str(value))
    print("Conditions: " + str(conditions))
```
